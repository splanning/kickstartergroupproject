```python
# find Undefined records using GBP as currency
ks2.loc[(ks2['state'] == 'undefined') & (ks2['currency'] == "GBP")]
```

```python
# fill in Country on ones with currency filled in and undefined state
ks2.loc[((ks2['state'] == 'undefined') & (ks2['currency'] == 'GBP')), 'country'] = 'GB'
```

```python
# do rest of countries
ks2.loc[((ks2['state'] == 'undefined') & (ks2['currency'] == 'USD')), 'country'] = 'US'
ks2.loc[((ks2['state'] == 'undefined') & (ks2['currency'] == 'AUD')), 'country'] = 'AU'
ks2.loc[((ks2['state'] == 'undefined') & (ks2['currency'] == 'CAD')), 'country'] = 'CA'
ks2.loc[((ks2['state'] == 'undefined') & (ks2['currency'] == 'SEK')), 'country'] = 'SE'
ks2.loc[((ks2['state'] == 'undefined') & (ks2['currency'] == 'DKK')), 'country'] = 'DK'
ks2.loc[((ks2['state'] == 'undefined') & (ks2['currency'] == 'NZD')), 'country'] = 'NZ'
ks2.loc[((ks2['state'] == 'undefined') & (ks2['currency'] == 'NOK')), 'country'] = 'NO'
ks2.loc[((ks2['state'] == 'undefined') & (ks2['currency'] == 'CHF')), 'country'] = 'CH'
```

```python
# Dropping undefined / EUR currency records - can't fill in country becuase it could be many
ks3 = ks2.drop(ks2[(ks2['state'] == 'undefined') & (ks2['currency'] == 'EUR')].index)
```

```python
#filling in country for rest of blank countries, except for EUR

ks3.loc[((ks3['country'] == 'N,0"') & (ks3['currency'] == 'GBP')), 'country'] = 'GB'
ks3.loc[((ks3['country'] == 'N,0"') & (ks3['currency'] == 'USD')), 'country'] = 'US'
ks3.loc[((ks3['country'] == 'N,0"') & (ks3['currency'] == 'AUD')), 'country'] = 'AU'
ks3.loc[((ks3['country'] == 'N,0"') & (ks3['currency'] == 'CAD')), 'country'] = 'CA'
ks3.loc[((ks3['country'] == 'N,0"') & (ks3['currency'] == 'SEK')), 'country'] = 'SE'
ks3.loc[((ks3['country'] == 'N,0"') & (ks3['currency'] == 'DKK')), 'country'] = 'DK'
ks3.loc[((ks3['country'] == 'N,0"') & (ks3['currency'] == 'NZD')), 'country'] = 'NZ'
ks3.loc[((ks3['country'] == 'N,0"') & (ks3['currency'] == 'NOK')), 'country'] = 'NO'
ks3.loc[((ks3['country'] == 'N,0"') & (ks3['currency'] == 'CHF')), 'country'] = 'CH'
```

```python
# dropping EUR no country records
ks4 = ks3.drop(ks3[(ks3['country'] == 'N,0"') & (ks3['currency'] == 'EUR')].index)
```

```python
ks4['country'].unique()
```

```python
# find Undefined records where amount pledged is higher than goal
ks4.loc[(ks4['state'] == 'undefined') & (ks4['pledged'] >= ks4['goal'])]
```

```python
# change state to successful on ones with undefined state and pledged higher than goal
ks4.loc[((ks4['state'] == 'undefined') & (ks4['pledged'] >= ks4['goal'])), 'state'] = 'successful'
```

```python
# and vice versa
ks4.loc[((ks4['state'] == 'undefined') & (ks4['pledged'] <= ks4['goal'])), 'state'] = 'failed'
```

```python
ks4.loc[(ks4['state'] == 'undefined')]
```

```python
# find records where successful but no known backers
ks4.loc[(ks4['state'] == 'successful') & (ks4['backers'] == 0)]
```

```python
# dropping successful but no known backers records
ks5 = ks4.drop(ks4[(ks4['state'] == 'successful') & (ks4['backers'] == 0)].index)
```

```python
ks5.loc[(ks5['state'] == 'successful') & (ks5['backers'] == 0)]
```

```python
# change state to successful on ones with live state and pledged higher than goal
ks5.loc[((ks5['state'] == 'live') & (ks5['pledged'] >= ks5['goal'])), 'state'] = 'successful'
```

```python
# dropping live records with unpredictable future
ks6 = ks5.drop(ks5[(ks5['state'] == 'live')].index)
```

```python
# dropping suspended records - no idea why they were suspended
ks7 = ks6.drop(ks6[(ks6['state'] == 'suspended')].index)
```

```python
# dropping low goal <100 USD records - too small to care to market to them
# not dropping any at high end - all appear to be legit
ks8 = ks7.drop(ks7[(ks7['usd_goal_real'] <= 100)].index)
```
